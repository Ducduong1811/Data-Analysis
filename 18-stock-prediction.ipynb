```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
import plotly.graph_objects as go
import plotly.express as px
```

```python
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1

d2 = date.today() - timedelta(days=365)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

```python
data = yf.download(
    'GOOG',
    start=start_date,
    end=end_date,
    progress=False
)

data["Date"] = data.index
df1 = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
df1.reset_index(drop=True, inplace=True)
df1.head()
```

```python
figure = go.Figure(
    data=[go.Candlestick(
        x=df1["Date"], open=df1["Open"], high=df1["High"], low=df1["Low"], close=df1["Close"]
    )]
)

figure.update_layout(title = "Google Stock Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

```python
df1.corr()
```

```python
from autots import AutoTS
model = AutoTS(forecast_length = 5, frequency = 'infer', ensemble = 'simple')
model = model.fit(df1, date_col = 'Date', value_col = 'Close', id_col = None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
